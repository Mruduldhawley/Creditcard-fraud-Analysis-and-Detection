# Upsampling

```python
from sklearn.utils import resample

not_fraud=X[X.Class==0]
fraud=X[X.Class==1]

#upsampling minority
fraud_upsampling=resample(fraud,replace=True,n_samples=len(not_fraud),random_state=101)

#combine minority and upsample minority
upsampled = pd.concat([not_fraud,fraud_upsampling])

upsampled.Class.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression

y_train = upsampled.Class
X_train = upsampled.drop('Class',axis=1)

upsampled = LogisticRegression(solver='liblinear').fit(X_train,y_train)

upsampled_pred = upsampled.predict(X_test)
```

```python
accuracy_score(y_test,upsampled_pred)
```

```python
from sklearn.metrics import classification_report as cr
```

```python
print('Logistic Regression Report')
print("______"*10)
print(cr(y_test,upsampled_pred))
```

```python

print('Confusion Matrix For Logistic Regression')
print('...'*10)
cm=metrics.confusion_matrix(y_test,upsampled_pred)
df_cm=pd.DataFrame(cm)
plt.figure(figsize=(7,5))
sns.heatmap(df_cm,annot=True)
pd.DataFrame(metrics.confusion_matrix(y_test,upsampled_pred))
```
